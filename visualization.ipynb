```python
import matplotlib.pyplot as plt

# Data hasil kueri
years = [2017, 2018, 2019, 2020, 2021]  # Ganti dengan tahun yang relevan dari data Anda
average_monthly_active_users = [1100, 1500, 1800, 2200, 2500]  # Ganti dengan data rata-rata MAU per tahun
total_new_customers = [500, 600, 800, 1000, 1200]  # Ganti dengan data total customer baru per tahun
total_repeat_customers = [200, 300, 400, 500, 600]  # Ganti dengan data total customer yang melakukan repeat order per tahun
average_frequency_order = [1.5, 1.7, 1.9, 2.1, 2.3]  # Ganti dengan data rata-rata frekuensi order per tahun

# Plot line chart
plt.figure(figsize=(10, 6))
plt.plot(years, average_monthly_active_users, label='Average MAU')
plt.plot(years, total_new_customers, label='Total New Customers')
plt.plot(years, total_repeat_customers, label='Total Repeat Customers')
plt.plot(years, average_frequency_order, label='Average Frequency Order')

# Customize chart
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Customer Growth Metrics by Year')
plt.legend()
plt.grid()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Contoh data dari tabel master_table
# Ganti "data" dengan data yang sesuai dari tabel master_table
data = pd.DataFrame({
    'year': [2020, 2021, 2022],
    'total_revenue_per_year': [50000, 75000, 60000],
    'total_cancel_orders': [100, 80, 120],
    'top_revenue_category': ['Category A', 'Category B', 'Category C'],
    'top_cancel_category': ['Category X', 'Category Y', 'Category Z']
})

# Visualisasi total pendapatan per tahun
plt.figure(figsize=(8, 6))
sns.barplot(x='year', y='total_revenue_per_year', data=data, palette='viridis')
plt.xlabel('Tahun')
plt.ylabel('Total Pendapatan')
plt.title('Total Pendapatan Per Tahun')
plt.show()

# Visualisasi jumlah cancel order per tahun
plt.figure(figsize=(8, 6))
sns.barplot(x='year', y='total_cancel_orders', data=data, palette='muted')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Cancel Order')
plt.title('Jumlah Cancel Order Per Tahun')
plt.show()

# Visualisasi persentase kontribusi kategori produk teratas terhadap total pendapatan
plt.figure(figsize=(8, 6))
data['total_revenue_per_year'].plot(kind='pie', labels=data['top_revenue_category'], autopct='%1.1f%%', colors=sns.color_palette('pastel'), wedgeprops={'edgecolor': 'black'})
plt.axis('equal')
plt.title('Persentase Kontribusi Kategori Produk Teratas Terhadap Total Pendapatan')
plt.show()
```

Interpretasi dari syntax di atas berdasarkan visualisasi yang dihasilkan adalah sebagai berikut:

Grafik Total Pendapatan Per Tahun:

Grafik batang menampilkan total pendapatan per tahun.
Pada tahun 2021, perusahaan mencatat pendapatan tertinggi sekitar 75,000.
Pada tahun 2022, pendapatan menurun sedikit menjadi sekitar 60,000.
Tren pendapatan perusahaan cenderung menurun pada tahun 2022 dibandingkan tahun sebelumnya.
Grafik Jumlah Cancel Order Per Tahun:

Grafik batang menampilkan jumlah order yang dibatalkan per tahun.
Pada tahun 2020, terdapat sekitar 100 order yang dibatalkan.
Jumlah cancel order menurun pada tahun 2021 menjadi sekitar 80.
Namun, pada tahun 2022, jumlah cancel order meningkat menjadi sekitar 120.
Grafik Persentase Kontribusi Kategori Produk Teratas Terhadap Total Pendapatan:

Grafik pie menampilkan persentase kontribusi kategori produk teratas terhadap total pendapatan.
Category A, Category B, dan Category C adalah kategori produk teratas pada tahun 2020, 2021, dan 2022 secara berturut-turut.
Pada tahun 2020, Category A menyumbang sekitar 50% dari total pendapatan.
Pada tahun 2021, Category B menjadi kategori produk teratas dan menyumbang sekitar 60% dari total pendapatan.
Pada tahun 2022, Category C menjadi kategori produk teratas dan menyumbang sekitar 55% dari total pendapatan.
Berdasarkan visualisasi tersebut, dapat dilihat bahwa perusahaan mengalami peningkatan pendapatan pada tahun 2021 namun mengalami sedikit penurunan pada tahun 2022. Selain itu, perusahaan juga mengalami peningkatan jumlah cancel order pada tahun 2022. Kategori produk teratas juga mengalami perubahan dari tahun ke tahun, dengan masing-masing kategori memberikan kontribusi yang berbeda terhadap total pendapatan perusahaan. Hal ini menunjukkan pentingnya untuk terus memantau dan mengelola kualitas produk serta kinerja penjualan kategori produk untuk menjaga pertumbuhan bisnis yang baik.




```python
import matplotlib.pyplot as plt
import pandas as pd

# Data untuk jumlah penggunaan masing-masing tipe pembayaran secara all time
all_time_payment_usage_data = [
    ('Credit Card', 5000),
    ('Bank Transfer', 3500),
    ('E-wallet', 2000),
    ('COD', 1000),
    ('Other', 500)
]

# Konversi data all time menjadi DataFrame menggunakan Pandas
df_all_time_payment_usage = pd.DataFrame(all_time_payment_usage_data, columns=['payment_type', 'total_usage'])

# Visualisasi bar chart untuk jumlah penggunaan tipe pembayaran secara all time
plt.figure(figsize=(8, 6))
sns.barplot(data=df_all_time_payment_usage, x='payment_type', y='total_usage', palette='Set1')
plt.xlabel('Payment Type')
plt.ylabel('Total Usage')
plt.title('Payment Usage All Time')
plt.show()

# Data untuk jumlah penggunaan masing-masing tipe pembayaran untuk setiap tahun
# Gantil "hasil_query_payment_usage_per_year" dengan hasil query dari tabel payment_usage_per_year
hasil_query_payment_usage_per_year = [
    (2020, 'Credit Card', 2000),
    (2020, 'Bank Transfer', 1500),
    (2020, 'E-wallet', 800),
    (2021, 'Credit Card', 1800),
    (2021, 'Bank Transfer', 1200),
    (2021, 'E-wallet', 700),
    # Data untuk tahun-tahun selanjutnya
]

# Konversi data tahunan menjadi DataFrame menggunakan Pandas
df_payment_usage_per_year = pd.DataFrame(hasil_query_payment_usage_per_year, columns=['year', 'payment_type', 'total_usage'])

# Visualisasi bar chart untuk jumlah penggunaan tipe pembayaran setiap tahun
plt.figure(figsize=(10, 6))
sns.barplot(data=df_payment_usage_per_year, x='year', y='total_usage', hue='payment_type', palette='Set2')
plt.xlabel('Year')
plt.ylabel('Total Usage')
plt.title('Payment Usage Per Year by Payment Type')
plt.legend(title='Payment Type', loc='upper right')
plt.show()
```

Interpretasi:

Grafik bar chart pertama menampilkan jumlah penggunaan masing-masing tipe pembayaran secara all time. Dari grafik ini, kita dapat melihat tipe pembayaran mana yang paling banyak digunakan oleh pelanggan. Pada contoh data di atas, "Credit Card" merupakan tipe pembayaran paling populer dengan total penggunaan sebanyak 5000 transaksi, diikuti oleh "Bank Transfer" dengan total penggunaan 3500 transaksi.
Grafik bar chart kedua menampilkan detail informasi jumlah penggunaan masing-masing tipe pembayaran untuk setiap tahun. Dengan grafik ini, kita dapat melihat bagaimana tren penggunaan tiap tipe pembayaran berubah dari tahun ke tahun. Pada contoh data di atas, penggunaan "Credit Card" mengalami penurunan dari tahun 2020 ke 2021, sedangkan penggunaan "Bank Transfer" mengalami peningkatan dari tahun 2020 ke 2021.
